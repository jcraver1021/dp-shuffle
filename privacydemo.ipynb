# Privacy Demonstration of Shuffling
If the client response order is lost, a great deal of privacy is gained.

```python
import longitudinal
import test

# Above are the modules constructed for this project
# longitudinal contains the client and server, as well as some methods for vector manipulation specific to this domain
# test contains an instance which encapsulates the entire environment, including running and stats, and some test suites

%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Utility to print reports
def print_reports(reports):
    n = len(reports[0])
    d = len(reports)
    for i in range(n):
        row = []
        for t in range(d):
            r = reports[t][i]
            if r:
                row.append(str(r[2]))
            else:
                row.append('X')
        print(','.join(row))
```

```python
# Make an instance to test with
rep = 20
n = 128
d = 16
k = 4
eps = 0.4
instance = test.Instance(n, d, k, eps)
```

```python
instance.run(True, False)
print_reports(instance.reports)
instance.write_reports("log_normal.csv")
```

```python
instance.run(True, True)
print_reports(instance.reports)
instance.write_reports("log_shuffled.csv")
```

```python
# Make a new instance to test with (will fix client parameters)
rep = 1024*64
n = 1024
d = 8
k = 1
eps = 0.4
instance = test.Instance(n, d, k, eps)

# Redefine clients to fix secret bits
# Only the 1st client will actually make a change; the rest will be static
#X = []
dX = []
clients = []
for i in range(n):
    if i == 0:
        dx = np.array([1] + [0] * (d - 1))
    else:
        dx = np.zeros(d)
    #x = longitudinal.compute_x(dx)
    dX.append(dx)
    #X.append(x)
    clients.append(longitudinal.Client(dx))
instance.clients = clients
#instance.X = np.array(X)
instance.dX = np.array(dX)

# Prepare the space where answers will be stored
x = np.arange(d)
y = np.zeros(d)
```

```python
def plot_progress(rep):
    plt.figure(dpi=250)
    plt.bar(x, y)
    plt.xlabel('Time')
    plt.ylabel('Total Reported Changes')
    plt.xticks(np.arange(0, d, 1))
    plt.title("Sum of Reports over 2^%d Epochs" % int(np.log2(rep)))
    plt.show()
```

```python
# Sum all the reported values
for i in range(rep + 1):
    y += instance.run(True, True)
    if i > 0 and i & (i-1) == 0:
        plot_progress(i)
```

```python
plot_progress(rep)
```
