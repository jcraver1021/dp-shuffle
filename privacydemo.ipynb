```python
# Utility to print reports
def print_reports(reports):
    n = len(reports[0])
    d = len(reports)
    for i in range(n):
        row = []
        for t in range(d):
            r = reports[t][i]
            if r:
                row.append("%d/%d/%d" %(r[0], r[1], r[2]))
            else:
                row.append("X/X/X")
        print(','.join(row))
```

```python
# Make an instance to test with
rep = 20
n = 1024
d = 32
k = 4
eps = 0.4
instance = test.Instance(n, d, k, eps)
```

```python
instance.run(True, False)
print_reports(instance.reports)
```

```python
instance.run(True, True)
print_reports(instance.reports)
```
